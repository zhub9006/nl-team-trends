# NL Team Trends Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('nl_team_wins.csv')
print(df.head())
print(f'Records: {len(df)}')
```

```python
teams = ['Los Angeles Dodgers', 'Atlanta Braves', 'St. Louis Cardinals', 'Philadelphia Phillies', 'San Francisco Giants', 'Chicago Cubs']
fig, ax = plt.subplots(figsize=(12, 6))
for team in teams:
    td = df[df['Team'] == team]
    ax.plot(td['Year'], td['WinPct'], marker='o', label=team)
ax.set_xlabel('Season')
ax.set_ylabel('Win %')
ax.set_title('NL Team Win % Trends (2011-2023)')
ax.legend()
plt.tight_layout()
plt.savefig('plots/win_trends.png', dpi=150)
plt.show()
```
